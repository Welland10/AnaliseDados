## Integração com o banco de Dados

#### Logar no banco de dados da empresa com as permissões
#### Extrair o arquivo para acpompanhamento das vendas
#### após salvar a prova da analise pronta, trocarei os numeros de servidores e ip para manter o direito a segurança da informação e proteção dos dados

```python
# Conectar ao banco de dados SQL Server
conn_str = (
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    r'SERVER=servidor;'
    r'DATABASE=banco de dados;'
    r'UID= usuario;'
    r'PWD=minhasenha' #troquei usuario e senha, servidor e data base por segurança
)
conn = pyodbc.connect(conn_str)

# Obter o mês atual
hoje = datetime.today()
ano_atual = hoje.year
mes_atual = hoje.month

# Consulta SQL para extrair pedidos do mês atual
query = f"""
SELECT 
    tp_fat AS 'Tipo de Fat'
    nr_pedcli AS 'Pedido',
    dt_ped_emis AS 'Data Pedido',
    dt_fat_emis AS 'Data Fat',
    ds_familia AS 'Familia',
    cod_item_datasul AS 'Item',
    ds_item AS 'Descrição Item',
    ped_status AS 'Status Pedido',
    ds_cli AS 'Cliente',
    ds_cidade AS 'Cidade',
    ds_estado AS 'UF',
    qtd AS 'Qtde',
    vl_liq AS 'Valor Total'
FROM dbo.DS_BASEVENDA
WHERE YEAR(DataPedido) = {ano_atual} AND MONTH(DataPedido) = {mes_atual} AND tp_fat = 'VENDA'
"""

# Ler os dados para um DataFrame
df = pd.read_sql(query, conn)

# Fechar a conexão
conn.close()
```

### Gerando o grafico para mostrar as vendas por dia

```python
vendas_por_mes = df.groupby('Data Pedido')['Valor Total'].sum()
plt.figure(figsize=(10, 6))
ax = vendas_por_mes.plot(kind='bar', color='blue')
plt.title('Vendas por Dia')
plt.xlabel('Dia')
plt.ylabel('Total de Vendas')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', 
                xy=(p.get_x() + p.get_width() / 2, height), 
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')
plt.show()
```

#### Gráfico com o Valor Total por Status de faturamento

```python
df['Venda Faturada'] = df['Status Pedido'].apply(lambda x: 'FATURADO' if x == 'FATURADO' else 'NAO FATURADO')
vendas_totais = df.groupby('Venda Faturada')['Valor Total'].sum()

# Criar o gráfico de barras
plt.figure(figsize=(8, 8))
ax = vendas_totais.plot(kind='bar', color=['skyblue', 'salmon'])  # Definir cores para as barras (opcional)
plt.title('Venda Total por Status de Faturamento')
plt.xlabel('Status de Faturamento')
plt.ylabel('Valor Total')
plt.xticks(rotation=0)

# Adicionar rótulos de dados em cima de cada barra
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',  # Rótulo com 2 casas decimais
                (p.get_x() + p.get_width() / 2, p.get_height()),  # Posição do rótulo
                ha='center', va='bottom')  # Alinhamento horizontal e vertical

plt.tight_layout()
plt.show()
```

### Gráfico com os percentuais de vendas faturadas e não faturadas

```python
status_vendas = df.groupby('Status Pedido')['Valor Total'].sum()
status_vendas = status_vendas[status_vendas >= 0].dropna() 
plt.figure(figsize=(8, 8)) 
plt.pie(status_vendas, 
        labels=status_vendas.index, 
        autopct='%1.1f%%',            
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],  
        startangle=140,               
        wedgeprops={'edgecolor': 'black'})  
plt.title('Distribuição de Vendas por Status de Pedido')
plt.show()
```

### Grafico com os 10 + Clientes

```python
# Ordena o DataFrame pelo valor de vendas em ordem decrescente
df_sorted = df.groupby('Cliente')['Valor Total'].sum().reset_index()
df_sorted = df_sorted.sort_values(by='Valor Total', ascending=False).head(10)

# Cria o gráfico de barras
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Valor Total', y='Cliente', data=df_sorted, hue='Cliente', dodge=False, palette='viridis', legend=False)

# Adiciona valores nos rótulos
for index, value in enumerate(df_sorted['Valor Total']):
    plt.text(value, index, f'{value:.2f}', color='black', va="center", ha="left")

# Adiciona título e rótulos
plt.title('Top 10 Clientes por Valor de Compras')
plt.xlabel('Valor das Vendas')
plt.ylabel('Cliente')

# Mostra o gráfico
plt.show()
```

### Grafico Com os 10 + Itens vendidos

```python
# Ordena o DataFrame pelo valor de vendas em ordem decrescente
df_sorted = df.groupby('Descrição Item')['Valor Total'].sum().reset_index()
df_sorted = df_sorted.sort_values(by='Valor Total', ascending=False).head(10)

# Cria o gráfico de barras
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Valor Total', y='Descrição Item', data=df_sorted, hue='Descrição Item', dodge=False, palette='viridis', legend=False)

# Adiciona valores nos rótulos
for index, value in enumerate(df_sorted['Valor Total']):
    plt.text(value, index, f'{value:.2f}', color='black', va="center", ha="left")

# Adiciona título e rótulos
plt.title('Top 10 Itens por Valor de Venda')
plt.xlabel('Valor das Vendas')
plt.ylabel('Itens')

# Mostra o gráfico
plt.show()
```

## Mostrar os Graficos todos de uma vez

```python
# Pré-processamento dos dados
df['Venda Faturada'] = df['Status Pedido'].apply(lambda x: 'FATURADO' if x == 'FATURADO' else 'NAO FATURADO')
vendas_totais = df.groupby('Venda Faturada')['Valor Total'].sum()
status_vendas = df.groupby('Status Pedido')['Valor Total'].sum()
status_vendas = status_vendas[status_vendas >= 0].dropna()
vendas_por_mes = df.groupby('Data Pedido')['Valor Total'].sum()
df_sorted_clientes = df.groupby('Cliente')['Valor Total'].sum().reset_index().sort_values(by='Valor Total', ascending=False).head(10)
df_sorted_itens = df.groupby('Descrição Item')['Valor Total'].sum().reset_index().sort_values(by='Valor Total', ascending=False).head(10)

# Criando subplots
fig, axes = plt.subplots(5, 1, figsize=(50, 70))

# Gráfico 1: Venda Total por Status de Faturamento
ax1 = vendas_totais.plot(kind='bar', color=['skyblue', 'salmon'], ax=axes[0])
ax1.set_title('Venda Total por Status de Faturamento')
ax1.set_xlabel('Status de Faturamento')
ax1.set_ylabel('Valor Total')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0)
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Gráfico 2: Distribuição de Vendas por Status de Pedido
axes[1].pie(status_vendas, labels=status_vendas.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], startangle=260, wedgeprops={'edgecolor': 'black'})
axes[1].set_title('Distribuição de Vendas por Status de Pedido')

# Gráfico 3: Vendas por Dia
ax3 = vendas_por_mes.plot(kind='bar', color='blue', ax=axes[2])
ax3.set_title('Vendas por Dia')
ax3.set_xlabel('Dia')
ax3.set_ylabel('Total de Vendas')
for p in ax3.patches:
    height = p.get_height()
    ax3.annotate(f'{height:.2f}', xy=(p.get_x() + p.get_width() / 2, height), xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

# Gráfico 4: Top 10 Clientes por Valor de Compras
sns.barplot(x='Valor Total', y='Cliente', data=df_sorted_clientes, hue='Cliente', palette='viridis', ax=axes[3], dodge=False, legend=False)
axes[3].set_title('Top 10 Clientes por Valor de Compras')
axes[3].set_xlabel('Valor das Vendas')
axes[3].set_ylabel('Cliente')
for index, value in enumerate(df_sorted_clientes['Valor Total']):
    axes[3].text(value, index, f'{value:.2f}', color='black', va="center", ha="left")

# Gráfico 5: Top 10 Itens por Valor de Venda
sns.barplot(x='Valor Total', y='Descrição Item', data=df_sorted_itens, hue='Descrição Item', palette='viridis', ax=axes[4], dodge=False, legend=False)
axes[4].set_title('Top 10 Itens por Valor de Venda')
axes[4].set_xlabel('Valor das Vendas')
axes[4].set_ylabel('Itens')
for index, value in enumerate(df_sorted_itens['Valor Total']):
    axes[4].text(value, index, f'{value:.2f}', color='black', va="center", ha="left")

plt.tight_layout()
plt.show()


```

## Salvar os gráficos em PDF para que o Analista possa enviar por e-mail ao gerente

```python
# Obter a data atual e formatar para o nome do arquivo
data_atual = datetime.now().strftime("%d.%m.%Y")  # Formata a data no formato DD.MM.AAAA
nome_arquivo = f"Vendas_San_Pio_Ate_{data_atual}.pdf"

# Definir o caminho para a pasta "Downloads" com a data atual no nome do arquivo
caminho_downloads = os.path.join(os.path.expanduser("~"), "Downloads", nome_arquivo)


# Definir o caminho para a pasta "Downloads"
#caminho_downloads = os.path.join(os.path.expanduser("~"), "Downloads", "graficos_gerados.pdf")

# Pré-processamento dos dados 
df['Venda Faturada'] = df['Status Pedido'].apply(lambda x: 'FATURADO' if x == 'FATURADO' else 'NAO FATURADO')
vendas_totais = df.groupby('Venda Faturada')['Valor Total'].sum()
status_vendas = df.groupby('Status Pedido')['Valor Total'].sum()
status_vendas = status_vendas[status_vendas >= 0].dropna()
vendas_por_mes = df.groupby('Data Pedido')['Valor Total'].sum()
df_sorted_clientes = df.groupby('Cliente')['Valor Total'].sum().reset_index().sort_values(by='Valor Total', ascending=False).head(10)
df_sorted_itens = df.groupby('Descrição Item')['Valor Total'].sum().reset_index().sort_values(by='Valor Total', ascending=False).head(10)

# Salvar gráficos no PDF na pasta Downloads
with PdfPages(caminho_downloads) as pdf:
    # Gráfico 1: Venda Total por Status de Faturamento
    plt.figure(figsize=(10, 6))
    ax1 = vendas_totais.plot(kind='bar', color=['skyblue', 'salmon'])
    plt.title('Venda Total por Status de Faturamento')
    plt.xlabel('Status de Faturamento')
    plt.ylabel('Valor Total')
    plt.xticks(rotation=0)
    for p in ax1.patches:
        plt.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
    pdf.savefig()
    plt.close()

    # Gráfico 2: Distribuição de Vendas por Status de Pedido
    plt.figure(figsize=(10, 6))
    plt.pie(status_vendas, labels=status_vendas.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], startangle=140, wedgeprops={'edgecolor': 'black'})
    plt.title('Distribuição de Vendas por Status de Pedido')
    pdf.savefig()
    plt.close()

    # Gráfico 3: Vendas por Dia
    plt.figure(figsize=(10, 6))
    ax3 = vendas_por_mes.plot(kind='bar', color='blue')
    plt.title('Vendas por Dia')
    plt.xlabel('Dia')
    plt.ylabel('Total de Vendas')
    plt.xticks(fontsize=8)  
    plt.yticks(fontsize=8)
    plt.tight_layout(pad=2)
    for p in ax3.patches:
        height = p.get_height()
        plt.annotate(f'{height:.2f}', xy=(p.get_x() + p.get_width() / 2, height), xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')
    pdf.savefig()
    plt.close()

    # Gráfico 4: Top 10 Clientes por Valor de Compras
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Valor Total', y='Cliente', data=df_sorted_clientes, hue='Cliente', palette='viridis', dodge=False, legend=False)
    plt.title('Top 10 Clientes por Valor de Compras')
    plt.xlabel('Valor das Vendas')
    plt.ylabel('Cliente')
    plt.xticks(fontsize=8)  
    plt.yticks(fontsize=8)
    plt.tight_layout(pad=2)
    for index, value in enumerate(df_sorted_clientes['Valor Total']):
        plt.text(value, index, f'{value:.2f}', color='black', va="center", ha="left")
    pdf.savefig()
    plt.close()

    # Gráfico 5: Top 10 Itens por Valor de Venda
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Valor Total', y='Descrição Item', data=df_sorted_itens, hue='Descrição Item', palette='viridis', dodge=False, legend=False)
    plt.title('Top 10 Itens por Valor de Venda')
    plt.xlabel('Valor das Vendas')
    plt.ylabel('Itens')
    plt.xticks(fontsize=8)  
    plt.yticks(fontsize=8)
    plt.tight_layout(pad=2) 
    for index, value in enumerate(df_sorted_itens['Valor Total']):
        plt.text(value, index, f'{value:.2f}', color='black', va="center", ha="left")
    pdf.savefig()
    plt.close()

print(f"Gráficos salvos com sucesso em: {caminho_downloads}")
```
